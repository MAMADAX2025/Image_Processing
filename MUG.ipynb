```python
#Code Explanation. Jun 15, 2023
'''
- Open an image from the root directory.
- Generate a random position in the image.
- Extract [LUDR] values within a range of 64px from the specified position.
L = left, U = top, D = bottom, R = right
- For a given matrix, generate only the corresponding MMM values.
M = max, M = average, M = min

The MRPM matrix represents the sum of all LUDR-RGB matrices and their respective MMMs.
----------------------------------------------------------------------------------------------------------------
#?[A1 - A5]
+ These values are stored in the registry and will be used later to reconstruct the target object.
'''

# Importing Image from PIL package

from PIL import Image, ImageDraw

# creating a image object
im = Image.open('./libreta.png')
rgb_im = im.convert('RGB')
w, h = im.size
# using getpixel method

def Average(lst):
    return sum(lst) / len(lst)

def get_left(x, y):
    MRPM.append(("LEFT"))
    for position in range(c):
        coordinate = x,y = x-1, y
        try:
            r, g, b = rgb_im.getpixel((coordinate))
            
            RL.append(r)
            GL.append(g)
            BL.append(b)
            
        except:
            return None
        
    MRPM.append(("r", RL,min(RL), round(Average(RL)), max(RL)))
    MRPM.append(("g", GL,min(GL), round(Average(GL)), max(GL)))
    MRPM.append(("b", BL,min(BL), round(Average(BL)), max(BL)))
    
    MRM_LEFT.append(min(RL))
    MRM_LEFT.append(min(GL))
    MRM_LEFT.append(min(BL))
    
    MRM_LEFT.append(round(Average(RL)))
    MRM_LEFT.append(round(Average(GL)))
    MRM_LEFT.append(round(Average(BL)))
    
    MRM_LEFT.append(max(RL))
    MRM_LEFT.append(max(GL))
    MRM_LEFT.append(max(BL))
        
def get_up(x, y):
    MRPM.append(("UP"))
    for position in range(c):
        coordinate = x,y = x, y+1
        try:
            r, g, b = rgb_im.getpixel((coordinate))
            
            RU.append(r)
            GU.append(g)
            BU.append(b)
            
        except:
            return None
        
    MRPM.append(("r", RU,min(RU), round(Average(RU)), max(RU)))
    MRPM.append(("g", GU,min(GU), round(Average(GU)), max(GU)))
    MRPM.append(("b", BU,min(BU), round(Average(BU)), max(BU)))
    
    MRM_UP.append(min(RU))
    MRM_UP.append(min(GU))
    MRM_UP.append(min(BU))
    
    MRM_UP.append(round(Average(RU)))
    MRM_UP.append(round(Average(GU)))
    MRM_UP.append(round(Average(BU)))
    
    MRM_UP.append(max(RU))
    MRM_UP.append(max(GU))
    MRM_UP.append(max(BU))
    
def get_down(x, y):
    MRPM.append(("DOWN"))
    for position in range(c):
        coordinate = x,y = x, y-1
        try:
            r, g, b = rgb_im.getpixel((coordinate))
            
            RD.append(r)
            GD.append(g)
            BD.append(b)
            
        except:
            return None

    MRPM.append(("r", RD,min(RD), round(Average(RD)), max(RD)))
    MRPM.append(("g", GD,min(GD), round(Average(GD)), max(GD)))
    MRPM.append(("b", BD,min(BD), round(Average(BD)), max(BD)))
    
    MRM_DOWN.append(min(RD))
    MRM_DOWN.append(min(GD))
    MRM_DOWN.append(min(BD))
    
    MRM_DOWN.append(round(Average(RD)))
    MRM_DOWN.append(round(Average(GD)))
    MRM_DOWN.append(round(Average(BD)))
    
    MRM_DOWN.append(max(RD))
    MRM_DOWN.append(max(GD))
    MRM_DOWN.append(max(BD))
    
def get_right(x, y):
    MRPM.append(("RIGHT"))
    for position in range(c):
        coordinate = x,y = x+1, y
        try:
            r, g, b = rgb_im.getpixel((coordinate))
            
            RR.append(r)
            GR.append(g)
            BR.append(b)
            
        except:
            return None
        
    MRPM.append(("r", RR,min(RR), round(Average(RR)), max(RR)))
    MRPM.append(("g", GR,min(GR), round(Average(GR)), max(GR)))
    MRPM.append(("b", BR,min(BR), round(Average(BR)), max(BR)))
    
    MRM_RIGHT.append(min(RR))
    MRM_RIGHT.append(min(GR))
    MRM_RIGHT.append(min(BR))
    
    MRM_RIGHT.append(round(Average(RR)))
    MRM_RIGHT.append(round(Average(GR)))
    MRM_RIGHT.append(round(Average(BR)))
    
    MRM_RIGHT.append(max(RR))
    MRM_RIGHT.append(max(GR))
    MRM_RIGHT.append(max(BR))

def get_pixels(x,y):
    get_left(x, y)
    get_up(x,y)
    get_down(x,y)
    get_right(x,y)
```

```python
#Create the dict to use
import collections
def most_used_words(value):
    words = collections.defaultdict(int)
    for line in value:
        for word in str(line).split():
            words[word] += 1
    return sorted(words, key=words.get, reverse=True)

#Ordenando los valores de mayor uso ;)
uniq = [int(item) for item in most_used_words(RL)]
print(uniq)
```

```python
items = ["hi world!"]
def minimax(string):
    output = []

    # Prefijos: "h", "ho", "hol"
    for i in range(1, len(string)+1):
        #output.append(f'"{string[:i]}"')
        output.append(string[:i])

    # Sufijos: "ola", "ol", "a"
    for i in range(1, len(string)):
        output.append(string[i:])
        
    return " ".join(output)
```
